## 1.Python Libraries

```python
import importlib
 
required = ['numpy', 'matplotlib', 'astropy', 'scipy', 'ipywidgets', 'setuptools', 'pyds9', 'h5py', 'pandas', 'firefly']

missing = []
 
for name in required:
    try:
        importlib.import_module(name)
        print(f'OK        {name}')
    except Exception as e:
        print(f'MISSING  {name}  ({e})')
        missing.append(name)

missing
```

## 2. Install the missing libs
Only if `missing` is not empty.

```python
import sys, subprocess
 
for pkg in missing:
    subprocess.run([sys.executable, '-m', 'pip', 'install', pkg])
```

After installations, run the cell in Section 1 again to confirm that they all show as `OK`.
